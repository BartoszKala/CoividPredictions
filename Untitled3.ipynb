```python
import pandas as pd
import numpy as np
from statsmodels.tsa.arima.model import ARIMA

# UsuÅ„ ostatnie 30 dni z aktualnego zbioru danych
new_data = df2.iloc[:-30].copy()

# Dopasuj model ARIMA do nowego zbioru danych
model_arima = ARIMA(new_data['New_cases'], order=(9,0,2))
model_arima_fit = model_arima.fit()

# Przewiduj dla nastÄ™pnych 30 dni
predictions = model_arima_fit.forecast(steps=30)

# UtwÃ³rz DataFrame z przewidywaniami
last_date = new_data['Date'].max()
prediction_dates = pd.date_range(start=last_date + pd.Timedelta(days=1), periods=30)
predicted_data = pd.DataFrame({
    'Date': prediction_dates,
    'New_cases': predictions,
    'All_cases_cumulative': np.nan,
    'Deaths': np.nan,
    'All_deaths_cumulative': np.nan,
    'Recovered_(daily)': np.nan,
    'Recovered_(sum)': np.nan,
    'Active_cases': np.nan,
    'Quarantine': np.nan,
    'Supervision': np.nan
})

# Dodaj przewidywania do tabeli new_data
new_data_with_predictions = pd.concat([new_data, predicted_data], ignore_index=True)

# WyÅ›wietl nowe dane z przewidywaniami
print(new_data_with_predictions.tail(40))


# Upewnij siÄ™, Å¼e kolumna 'Date' w 'df2' jest typu datetime
df2['Date'] = pd.to_datetime(df2['Date'], format='%d.%m.%Y')

# Wybierz ostatnie 35 dni z obu DataFrame'Ã³w
last_35_days_predictions = new_data_with_predictions.iloc[-35:]
last_35_days_original = df2.iloc[-35:]

plt.figure(figsize=(12, 6))

# Wykres dla przewidywaÅ„
plt.plot(last_35_days_predictions['Date'], last_35_days_predictions['New_cases'], marker='o', linestyle='-', label='Predictions')

# Wykres dla oryginalnych danych
plt.plot(last_35_days_original['Date'], last_35_days_original['New_cases'], marker='x', linestyle='--', label='Original Data')

plt.xlabel('Date')
plt.ylabel('New Cases')
plt.title('New COVID-19 Cases Over Time (Last 35 Days)')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

```
